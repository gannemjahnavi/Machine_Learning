```python
import numpy as np
import pandas as pd
```

```python
def euclideanDistance(x1,x2):
  return np.sqrt(np.sum((x1-x2)**2))
```

```python
class KNN:
  def __init__(self,k):
    self.k=k

  def fit(self,x,y):
    self.xtrain=x
    self.ytrain=y
  
  def predict(self,x):
    ypred=[]
    for i in x:
      distances=[euclideanDistance(i,x_train) for x_train in self.xtrain]
      k_indices=np.argsort(distances)[:self.k]
      k_neighbours=[self.ytrain[i] for i in k_indices]
      ypred.append(max(k_neighbours,key=k_neighbours.count))
    return np.array(ypred)
```
